```python
titanic['Age'] = titanic['Age'].fillna(titanic['Age'].mean())
```

```python
titanic.drop('Cabin', axis=1, inplace=True)
```

```python
titanic['Sex'].value_counts()
```

```python
titanic['Embarked'].value_counts()
```

```python
sex = pd.get_dummies(titanic['Sex'], drop_first=True)
sex
```

```python
emb = pd.get_dummies(titanic['Embarked'])
emb
```

```python
titanic = pd.concat([titanic, sex,emb], axis=1)
```

```python
titanic
```

```python
titanic.drop(['PassengerId', 'Name', 'Sex', 'Ticket', 'Embarked'], axis=1, inplace=True)
```

```python
titanic.rename(columns={'male':'Gender'}, inplace=True)
```

```python
x = titanic.drop('Survived', axis=1)
y = titanic['Survived']
```

## Train Test Model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

## Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.svm import SVC
```

```python
model = LogisticRegression(max_iter=5000) #model building
```

```python
model.fit(x_train,y_train) #model trainning
```

```python
y_pre = model.predict(x_test) #model testing
y_pre
```

```python
y_test.values
```

## Model Evaluation

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pre)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pre))
```
